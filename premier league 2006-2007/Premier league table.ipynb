```python
import csv
dataset = list(csv.reader(open("results.csv")))
dataset
```

```python
# list the name of teams 
teams = []
for i in dataset[1:]:
    season = i[-1]
    if season=='2006-2007':
        h_team = i[0]
        a_team = i[1]
        if h_team not in teams:
            teams.append(h_team)
        if a_team not in teams:
            teams.append(a_team)
    else:
        break
print(teams)
```

```python
# now make table
freq = {}
table = [['CLUB','MP','W','D','L','GF','GA','GD','PTS']]
for i in teams:
    team = i
    for j in dataset[1:]:
        mp=0 
        w=0
        l=0
        d=0
        gf=0
        ga=0
        gd=0
        pts=0 
        home_team = j[0]
        away_team = j[1]
        home_goals = int(float(j[2]))
        away_goals = int(float(j[3]))
        goals_diff = home_goals - away_goals
        result = j[5]
        season = j[-1]
        if season=='2006-2007':
            if i==home_team:
                gf = home_goals
                ga =  away_goals
                mp=1
                if home_goals>away_goals:
                    w =1
                    pts=3
                    gd = gf - ga
                elif home_goals==away_goals:
                    d=1
                    pts=1
                    gd = 0
                else:
                    l=1
                    pts=0
                    gd = ga - gf
            elif i == away_team:
                mp=1
                gf = away_goals
                ga =  home_goals
                if away_goals>home_goals:
                    w =1
                    pts=3
                    gd = gf - ga
                elif home_goals==away_goals:
                    d=1
                    pts=1
                    gd = 0
                else:
                    l=1
                    pts=0
                    gd = ga - gf
            if i not in freq:
                freq[team] = [team,mp,w,d,l,gf,ga,gd,pts]
            else:
                freq[team][1]+=mp
                freq[team][2]+=w
                freq[team][3]+=d
                freq[team][4]+=l
                freq[team][5]+=gf
                freq[team][6]+=ga
                freq[team][7]+=gd
                freq[team][-1]+=pts 

        else:
            break   
            
for i in freq:
        freq[i][7] = freq[i][5] - freq[i][6]
```

```python
for i in freq.values():
    table.append(i)
```
